```python
#lr.adj_close_lin_reg(lr.df_2017, 5)
data = lr.adj_close_lin_reg(lr.df_2017, 10)

long_days = data.long_count.sum()
short_days = data.short_count.sum()

profit_col_name = 'w_' + str(10) + '_profit'


def short_or_long_trx(vector):
    if vector[0] is not None and vector[1] == 1:
        return 1
    else: 
        return 0
    
def short_or_long_profit(vector):
    if vector[1] == 1: 
        return vector[0]
    else:
        return 0

data['short_trx'] = data[[profit_col_name, 'short_count']].apply(short_or_long_trx, axis=1)
data['long_trx'] = data[[profit_col_name, 'long_count']].apply(short_or_long_trx, axis=1)

data['short_profit'] = data[[profit_col_name, 'short_trx']].apply(short_or_long_profit, axis=1)
data['long_profit'] = data[[profit_col_name, 'long_trx']].apply(short_or_long_profit, axis=1)

transactions = data[profit_col_name].count()

long_transactions = data.long_trx.sum()
short_transactions = data.short_trx.sum()

print(long_days)
print(short_days)
print(transactions)
print('Long Trx: ', long_transactions)
print('Short Trx: ', short_transactions)


long_profit_avg = data.long_profit.sum() / long_transactions
short_profit_avg = data.short_profit.sum() / short_transactions


long_avg_days = long_days / long_transactions
short_avg_days = short_days / short_transactions

print('Avg long profit: ', long_profit_avg)
print('Avg short profit: ', short_profit_avg)

print('# Days long avg: ', long_avg_days)
print('# Days short avg: ', short_avg_days)

print(data)
```
